#### **CONCAT** ####

```python
%sql SELECT CONCAT('a','b');
```

```python
%%sql 
SELECT CONCAT(author_fname,' ',author_lname) AS Complete_name FROM books;
```

#### **CONCAT_WS** ####

```python
%%sql 
SELECT CONCAT_WS (' - ',title	,author_fname,author_lname) AS Complete_name FROM books;
```

#### **SUBSTRING** ####

```python
%%sql
SELECT SUBSTRING(title,1,100) FROM books
```

```python
%%sql
SELECT SUBSTRING(title, 1) FROM books
```

```python
%%sql
SELECT SUBSTRING(title, -2) FROM books
```

#### **Combined funcions** ####

```python
%sql SELECT CONCAT(SUBSTRING(title,1,10),'...') FROM Books;
```

```python
%%sql
SELECT CONCAT(
    SUBSTRING(author_fname,1,1),
    '.',
    SUBSTRING(author_lname,1,1),
    '.')
AS Autor_initials 
FROM Books;
```

#### **Replace** ####

```python
%%sql
SELECT REPLACE('Hello World', 'Hell', '%$#@');
 
SELECT REPLACE('Hello World', 'l', '7');
 
SELECT REPLACE('Hello World', 'o', '0');
 
SELECT REPLACE('HellO World', 'o', '*');
 
SELECT
  REPLACE('cheese bread coffee milk', ' ', ' and ');
 
SELECT REPLACE(title, 'e ', '3') FROM books;
 
SELECT REPLACE(title, ' ', '-') FROM books;
```

#### **Reverse** ####

```python
%%sql
SELECT REVERSE('Hello World');
 
SELECT REVERSE('meow meow');
 
SELECT REVERSE(author_fname) FROM books;
 
SELECT CONCAT('woof', REVERSE('woof'));
 
SELECT CONCAT(author_fname, REVERSE(author_fname)) FROM books;
```

#### CHAR_Lenght ####

```python
%%sql
SELECT CHAR_LENGTH('Hello World');
 
SELECT CHAR_LENGTH(title) as length, title FROM books;
 
SELECT author_lname, CHAR_LENGTH(author_lname) AS 'length' FROM books;
 
SELECT CONCAT(author_lname, ' is ', CHAR_LENGTH(author_lname), ' characters long') FROM books;
```

#### **UPPER-LOWER** ####

```python
%%sql
SELECT UPPER('Hello World');
 
SELECT LOWER('Hello World');
 
SELECT UPPER(title) FROM books;
 
SELECT CONCAT('MY FAVORITE BOOK IS ', UPPER(title)) FROM books;
 
SELECT CONCAT('MY FAVORITE BOOK IS ', LOWER(title)) FROM books;
```
