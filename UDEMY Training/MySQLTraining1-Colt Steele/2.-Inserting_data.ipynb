#### **Inserta datos en la tabla** ####

````python
%%sql 
INSERT INTO cats(name,age)
VALUES ("bolita", 2);
````

#### **Insertar Datos Multiples** ####

````python
%%sql
INSERT INTO cats(name,age)
VALUES
("oreo",3),
("pocoyo",4),
("leon",5);
````

#### **Valores NULL** ####
- Nos permite decir si una parte de la tabla puede tener NULLS

````python
%%sql 
INSERT INTO cats (name,age)
VALUES ("tobias", NULL);
````

````python
%sql SELECT * FROM cats;
````

````python

%%sql
INSERT INTO cats (name, age)
VALUES (NULL, NULL);
````

````python
%sql SELECT * FROM cats;
````

````python
%%sql

CREATE TABLE cats_restricted (
    name VARCHAR(50) NOT NULL,
    age INT NOT NULL
);
````

````python
%%sql
SELECT name
FROM sqlite_master
WHERE type="table";
````

````python
%sql PRAGMA table_info(cats_restricted)
````

#### **DEFAULT VALUES** ####

````python
%%sql 
CREATE TABLE cats_deafault(
    name VARCHAR(50) DEFAULT "gato",
    age INT DEFAULT 0
)
````

````python
%%sql
SELECT name
FROM sqlite_master
WHERE type="table";
````

````python
%sql PRAGMA table_info(cats_default);
````

````python
%%sql
INSERT INTO cats_deafault (name,age)
VALUES (NULL, NULL);
````
