#### **DISTINCT** ####

```python
%sql SELECT DISTINCT author_fname, author_lname FROM books;
```

```python
%sql SELECT DISTINCT author_fname FROM books;
```

```python
%sql SELECT DISTINCT author_fname, title FROM books;
```

#### **ORDER BY** ####

```python
%sql SELECT DISTINCT author_fname, title FROM books ORDER BY author_fname, title, stock_quantity;
```

```python
%sql SELECT DISTINCT author_fname, title FROM books ORDER BY stock_quantity;
```

```python
%sql SELECT DISTINCT author_fname, title FROM books ORDER BY stock_quantity DESC;
```

##### ORDER BY NUMBER OF COLUMN #####

```python
%sql SELECT DISTINCT author_fname, title FROM books ORDER BY 2;
```

#### **LIMIT** ####

```python
%sql SELECT DISTINCT author_fname, title FROM books LIMIT 5;
```

```python
%sql SELECT DISTINCT author_fname, title, released_year FROM books ORDER BY released_year LIMIT 5;
```

```python
%sql SELECT DISTINCT author_fname, title, released_year FROM books ORDER BY released_year LIMIT 1,5;
```

#### **LIKE** ####

```python
%sql SELECT title FROM books WHERE title = 'White Noise'
```

```python
%sql SELECT title FROM books WHERE title LIKE '%mala%'
```
