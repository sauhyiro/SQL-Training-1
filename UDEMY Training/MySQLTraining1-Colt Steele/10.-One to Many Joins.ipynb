#### **CROSS JOINS** ####

```python
%%sql
SELECT CONCAT(customer_first_name,1) FROM customers WHERE customer_id=1;
```

```python
%sql SELECT * from customers, orders;
```

#### **INNER JOIN** ####

```python
%%sql
SELECT customers.customer_first_name, customers.customer_last_name, orders.order_date, SUM(orders.order_amount) AS total_order_amount FROM customers 
JOIN orders
    ON customers.customer_id = orders.customer_id
GROUP by customers.customer_id;
```

#### **LEFT JOIN** ####

```python
%%sql
SELECT * from customers
LEFT JOIN orders
    ON customers.customer_id = orders.customer_id;
```

#### **RIGHT JOIN** ####

```python
%%sql
SELECT * from customers
RIGHT JOIN orders
    ON customers.customer_id = orders.customer_id;
```

```python
%%sql
SELECT 
    customer_first_name, 
    customer_last_name, 
    IFNULL(SUM(orders.order_amount), 0) AS money_spent
FROM
    customers
        LEFT JOIN
    orders ON customers.customer_id = orders.customer_id
GROUP BY customer_first_name, customer_last_name;
```

#### **CASCADE DELETE** ####

- Si borramos un cliente que esta en una llave externa. dara error por que todavia existe en otra.

```python
%%sql
CREATE TABLE orders2(
    order_id INTEGER PRIMARY KEY AUTOINCREMENT,
    order_date DATETIME,
    order_amount DECIMAL(10,2),
    customer_id INTEGER,
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id) ON DELETE CASCADE
)
```
