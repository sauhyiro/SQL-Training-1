#### **GROUP BY** ####

```python
%%sql
SELECT COUNT(*) AS Book_number, author_lname FROM books
GROUP BY author_lname ORDER BY Book_number DESC
```

```python
%%sql
SELECT released_year, COUNT(*) AS num FROM books GROUP BY released_year ORDER BY num;
```

```python
%sql SELECT * FROM books GROUP by author_lname;
```

#### **MIN-MAX** ####

```python
%sql SELECT MIN(pages), title FROM Books;
```

```python
%sql SELECT MAX(pages), title FROM Books;
```

#### **SUBQUERIES** ####

```python
%sql SELECT title, pages FROM books ORDER BY pages DESC LIMIT 1;
```

```python
%%sql
SELECT * FROM books
WHERE pages = (SELECT MAX(pages) FROM books);
```

```python
%%sql
SELECT title, released_year FROM books 
WHERE released_year = (SELECT MIN(released_year) FROM books);
```

#### **GROUP SEVERAL COLUMS** ####

```python
%sql SELECT author_lname, author_fname, COUNT(*) FROM books GROUP BY author_lname, author_fname;
```

```python
%%sql
SELECT CONCAT(Author_fname, ' ', author_lname) AS author, COUNT(*)
FROM books
GROUP BY author;
```

#### **MIN / MAX / GROUPING** ####

```python
%%sql
SELECT author_lname, MIN(released_year) FROM books GROUP BY author_lname;
```

```python
%%sql
SELECT CONCAT(author_lname,' ', author_fname), COUNT(*), MIN(released_year), MAX(released_year) , MAX(pages) FROM books GROUP BY author_lname, author_fname ORDER BY author_lname desc;
```

#### **SUM** ####

```python
%%sql
SELECT author_lname, author_fname, SUM(pages) AS Total_pages 
FROM books
GROUP BY author_lname, author_fname
ORDER BY Total_pages DESC;
```

```python
%%sql
SELECT author_lname, COUNT(*) AS Book_count, SUM(pages) AS Total_pages
FROM books
GROUP BY author_lname;
```

#### **AVERAGE** ####

```python
%%sql
SELECT author_lname, COUNT(*) AS Book_count, AVG(pages) AS Average_pages
FROM books
GROUP BY author_lname;
```

```python
%%sql
SELECT 
    released_year, 
    AVG(stock_quantity), 
    COUNT(*) FROM books
GROUP BY released_year;
```
