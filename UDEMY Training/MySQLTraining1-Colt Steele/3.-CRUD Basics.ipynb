##### Working Table ####

````python
%%sql
CREATE TABLE cats(
    cat_id INT PRIMARY KEY,
    name VARCHAR(100),
    breed VARCHAR(100),
    age INT
);
````

````python
%%sql
INSERT INTO cats (cat_id,name,breed,age)
VALUES
(1,"ringo","taby",4),
(2,"cindy","maine coon",10),
(3, "dumbedore", "maine coon",11),
(4, "egg", "persian",4),
(5, "mysty", "taby",13),
(6, "george michel", "ragdol", 9),
(7, "jacson", "sphynx",7);
````

````python
%sql SELECT * FROM cats
````
